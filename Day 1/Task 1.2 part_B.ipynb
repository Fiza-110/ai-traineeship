```python
import math
import random
import numpy as np


def dot_product(a, b):
    if len(a) != len(b):
        raise ValueError("Vectors must have same length")

    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result


def magnitude(v):
    total = 0
    for x in v:
        total += x * x
    return math.sqrt(total)


def cosine_similarity(a, b):
    mag_a = magnitude(a)
    mag_b = magnitude(b)

    if mag_a == 0 or mag_b == 0:
        raise ValueError("Zero vector not allowed")

    return dot_product(a, b) / (mag_a * mag_b)


def mean(v):
    if len(v) == 0:
        raise ValueError("Empty vector")

    return sum(v) / len(v)


def variance(v):
    if len(v) == 0:
        raise ValueError("Empty vector")

    m = mean(v)
    return sum((x - m) ** 2 for x in v) / len(v)


# TESTING
TOL = 1e-9

for _ in range(100):
    size = random.randint(2, 20)

    a = [random.uniform(-100, 100) for _ in range(size)]
    b = [random.uniform(-100, 100) for _ in range(size)]

    assert abs(dot_product(a, b) - np.dot(a, b)) < TOL
    assert abs(magnitude(a) - np.linalg.norm(a)) < TOL

    if np.linalg.norm(a) != 0 and np.linalg.norm(b) != 0:
        cos_np = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
        assert abs(cosine_similarity(a, b) - cos_np) < TOL

    assert abs(mean(a) - np.mean(a)) < TOL
    assert abs(variance(a) - np.var(a)) < TOL

print("All tests passed successfully!")

print("Heloo")
```
